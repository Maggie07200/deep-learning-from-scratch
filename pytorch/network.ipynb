```python
#
def AND(x1,x2):
    w1,w2,theta = 0.5,0.5,0.8
    tmp = x1*w1 + x2*w2
    if tmp >= theta:
        return 1
    elif tmp < theta:
        return 0
    
```

```python
#utilize numpy to realize nand
import numpy as np
def NAND(x1,x2):
    x = np.array([x1,x2])
    w = np.array([-0.5,0.5])
    b = 0.8
    temp = np.sum(x*w)
    if temp:
        return 1
    else:
        return 0
```

```python
def OR(x1,x2):
    x = np.array([x1,x2])
    w = np.array([0.5,0.5])
    b = -0.2
    temp = np.sum(x*w)
    if temp:
        return 1
    else:
        return 0
```

```python
def XOR(x1,x2):
    s1 = NAND(x1,x2)
    s2 = OR(x1,x2)
    y = AND(s1,s2)
    return y
```

```python
def identity_function(x):
    return x
def sigmoid(x):
    y = 1 + np.exp(x)
    return 1/y
def init_network():
    network = {}
    network['w1'] = np.array([[0.1,0.2,0.3],[0.2,0.1,0.4]])
    network['b1'] = np.array([0.1,0.2,0.3])
    network['w2'] = np.array([[0.1,0.2],[0.2,0.1],[0.7,0.4]])
    network['b2'] = np.array([0.1,0.2])
    network['w3'] = np.array([[0.1,0.3],[0.1,0.4]])
    network['b3'] = np.array([0.1,0.3])
    return network

```

```python
# bulid forward
def forward(network,X):
    w1, w2, w3 = network['w1'], network['w2'], network['w3']
    b1,b2,b3=network['b1'],network['b2'],network['b3']
    A1 = np.dot(X,w1) + b1
    Z1 = sigmoid(A1)
    A2 = np.dot(Z1,w2) + b2
    Z2 = sigmoid(A2)
    A3 = np.dot(Z2,w3) + b3
    Z3 = identity_function(A3)
    return Z3
```

```python
network = init_network()
x = np.array([1.0,0.5])
y = forward(network,x)
print (y)
```
