learn how to use matplotlib to draw diagrams

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,6,0.1)
y = np.sin(x)
plt.plot(x,y)
plt.show
```

```python
# add cos
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
#draw the diagrams
plt.plot(x,y1,label = 'sin')
plt.plot(x,y2,linestyle = "--",label = "cos")
plt.xlabel('x')
plt.ylabel('y')
plt.title("sin and cos")
plt.legend()
plt.show()
```
