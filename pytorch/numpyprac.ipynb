```python
#importnumpy
import numpy as np
```

```python
# create a numpy array and do some basic calculumn
m = np.array([1,2,3])
n = np.array([4.0,5.0,6.0])
print(m+n)
print(m*n)
print(m/n)
print(m/2.0)
```

```python
#create a numpy matrix
A = np.array([[1,2],[3,4]])
print(A.shape)
print(A.dtype)
print(A)
```

```python
B = np.array([[4,5],[6,7]])
print(A+B)
print(A*B)
```

```python
# about broadcast
M = np.array([[1,2],[3,4]])
N = np.array([10,20])
N1 = np.array([[10],[20]])
print(M*N) 
print(M*N1)
```

```python
#how about A*B
A = np.array([[1,2],[3,4],[5,6]])
B = np.array([[1,2,3],[4,5,6]])
print(A@B)
print(np.dot(A,B))
print(np.matmul(A,B))
```

```python
#then about Transpose
A_T = A.T
print(A_T)
A_t = np.transpose(A)
print(A_t)
```

```python
#then comes to index
X = np.array([[1,2,5],[2,1,6],[4,4,89]])
print(X)
print(X[0])
print(X[1][2])
print(X[1,2])
```

```python
print(X[1,:])
print(X[:,0])
```

```python
print(X[:2,1:3])
```

```python
m = np.arange(len(X))
n = np.array([0,2,1])
print(X[m,n])
#相当于是去找坐标
```

```python
X = X.flatten()
print(X)
print(X[np.array([0,2,4])])
print(X>15)
print(X[X>15])
```

```python
# about sum beyond dimensions
A = np.array([[1,2],[3,4]])
print(A)
print(np.sum(A))
print(np.sum(A,axis = 0))
print(np.sum(A,axis=1))
```

```python
# some upgraded knowledge
a = np.array([[1,2],[3,4]])
b = np.array([[5,6]])
add = np.tile(b,(2,2))#represent复制几行几列
add1 = np.tile(a,(2,4))
print(add)
print(add1)
```

```python
#change dimensions
x = np.reshape(add,(4,2))
print(x)
```

```python
#delete dimensions
x = np.array([[[1,2,1],[1,2,1]]])
print(x.shape)
y = np.squeeze(x,0)
print(y)
print(y.shape)
```

```python
#add dimensions
z = np.expand_dims(y,1)
print(z)
print(z.shape)
```

```python
#some interesting functions
a = np.zeros((2,3))
b = np.ones((3,2))
c = np.eye((4))
d = np.random.random((2,1))
#attention two slice of ()
print(a)
print(b)
print(c)
print(d)
```
